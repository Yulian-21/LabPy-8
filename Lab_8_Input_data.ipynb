```
import matplotlib.pyplot as plt
import numpy as np


def get_dataset():
    np.random.seed(1)
    m = 400
    N = int(m / 2)  # number of points per class

    dimension = 2
    X = np.zeros((m, dimension))  # data matrix where each row is a single example
    Y = np.zeros((m, 1), dtype='uint8')  # labels vector (0 for red, 1 for blue)
    flower_ray = 4

    for j in range(2):
        indexs = range(N * j, N * (j + 1))
        t = np.linspace(j * 3.12, (j + 1) * 3.12, N) + np.random.randn(N) * 0.2
        r = flower_ray * np.sin(4 * t) + np.random.randn(N) * 0.2
        X[indexs] = np.c_[r * np.sin(t), r * np.cos(t)]
        Y[indexs] = j

    X = X.T
    Y = Y.T
    return X, Y


# if name == "main":
#     X, Y = get_dataset()
#     plt.scatter(X[0, :], X[1, :], c=Y[0], s=40, cmap=plt.cm.Spectral);
#     plt.show()

from sklearn.cluster import AffinityPropagation as AgglomerativeClustering
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
np.random.seed(1)
x1, y1 = get_dataset()
x = []
for i, n in enumerate(x1[0]):
    x.append([n, x1[1][i]])

xx = x1[0]
yy = x1[1]
em = AgglomerativeClustering(convergence_iter=2).fit([[-2,1],[0,3],[2,1],[2,-1],[0,-3],[-2,-1]])
em.cluster_centers_ = np.array([[-2,0.4],[0,2],[2,0.6],[2,-0.4],[0,-2],[-2,-1]])
aggloclust = em.predict(x)
labels = aggloclust
plt.scatter(xx,yy, c=labels, s=40, cmap=plt.cm.Spectral)
plt.show()

f = plt.figure()
f.add_subplot(2, 2, 1)

plt.show()
```
